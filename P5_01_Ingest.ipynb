```python
from pyspark.sql.functions import current_timestamp, lit

df_source_oil = spark.table(SOURCE_TABLE_OIL)
df_source_holidays = spark.table(SOURCE_TABLE_HOLIDAYS)
df_source_sales = spark.table(SOURCE_TABLE_SALES)
```

```python
df_source_oil = df_source_oil.withColumn("ingested_at", current_timestamp()).withColumn("run_id", lit(RUN_ID))
df_source_oil.write.format("delta").mode("overwrite").saveAsTable(BRONZE_OIL)
```

```python
df_source_holidays = df_source_holidays.withColumn("ingested_at", current_timestamp()).withColumn("run_id", lit(RUN_ID))
df_source_holidays.write.format("delta").mode("overwrite").saveAsTable(BRONZE_HOLIDAYS)
```

```python
df_source_sales = df_source_sales.withColumn("ingested_at", current_timestamp()).withColumn("run_id", lit(RUN_ID))
df_source_sales.write.format("delta").mode("overwrite").saveAsTable(BRONZE_SALES)
```
